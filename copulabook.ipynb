```python
import pandas as pd
import numpy as np
from shapiq_student.imputer import GaussianCopulaImputer
```

```python
data = {
    'age': [48.935913, 39.234323, 55.659901, 31.810637, 65.342336],
    'income': [399.161393, 364.225531, 406.475105, 341.276022, 414.347815]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('data.csv', index=False)
```

```python
df = pd.read_csv('data.csv')
print(df.head())
```

```python
X = df.values
mask = np.zeros_like(X, dtype=bool)
```

```python
X_train = X[:, 0].reshape(-1, 1)  # age als Feature
y_train = X[:, 1]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
def trained_model(X_input):
    # Falls X_input mehrere Features hat, wähle hier nur 'age' (Spalte 0) für Vorhersage
    X_age = X_input[:, 0].reshape(-1, 1)
    return model.predict(X_age)
```

```python
imputer = GaussianCopulaImputer(model=trained_model, data=X)

imputer.fit(X)

X_missing = X.copy()
mask = np.zeros_like(X_missing, dtype=bool)
mask[1, 1] = True
X_missing[1, 1] = np.nan

X_imputed = imputer.transform(X_missing, mask)

print("Originaldaten mit fehlenden Werten:")
print(X_missing)
print("\nDaten nach Imputation:")
print(X_imputed)
```
